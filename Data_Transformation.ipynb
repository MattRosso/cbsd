# Tranformation of the data

```python
import numpy as np 
import matplotlib.pyplot as plt 
import csv
with open('RAW_DDDT.CSV') as csv_file:
    csv_read=csv.reader(csv_file, delimiter=',')

import pandas as pd
data = pd.read_csv('RAW_DDDT.CSV')
df = pd.DataFrame(data)

Y = df['CONDITION']

e = []
for elem in df:
    e.append(elem)
```

```python
data["Condition"] = data.CONDITION.map({"H":0, "D":5})
data.drop("CONDITION", axis = 1,inplace = True)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = scaler.fit_transform(data)
df1_scaled = pd.DataFrame(X, columns=e)
df1_scaled
```

```python
honest = df1_scaled[df1_scaled['CONDITION'] == 0]
honest = honest.iloc[:,:12]
dishonest = df1_scaled[df1_scaled['CONDITION'] == 1]
dishonest = dishonest.iloc[:,:12]
honest
```

## Average Transformation

```python
model_dict = {"PSY": ["P1", "P3", "P5", "P10"], "NAR": ["N2", "N7", "N8", "N12"], "MAC": ["M4", "M6", "M9", "M11"]}
```

```python
x = honest.loc[:, model_dict['PSY']]
y = honest.loc[:, model_dict['NAR']]
z = honest.loc[:, model_dict['MAC']]

coord = np.array([x.mean(axis=1),y.mean(axis=1),z.mean(axis=1)]).T
```

```python
x = dishonest.loc[:, model_dict['PSY']]
y = dishonest.loc[:, model_dict['NAR']]
z = dishonest.loc[:, model_dict['MAC']]

coord1 = np.array([x.mean(axis=1),y.mean(axis=1),z.mean(axis=1)]).T
```

```python
from mpl_toolkits import mplot3d

x = coord[:,0]
y = coord[:,1]
z = coord[:,2]

x1 = coord1[:,0]
y1 = coord1[:,1]
z1 = coord1[:,2]

fig = plt.figure(figsize = (10, 10))
ax = plt.axes(projection ="3d")

ax.scatter3D(x, y, z, c='b', marker='o', label='Honest')
ax.scatter3D(x1, y1, z1, c='r', marker='o', label='Disonest')

ax.set_xlabel('Psychopathy')
ax.set_ylabel('Narcissism')
ax.set_zlabel('Machiavellianism')
plt.title('Components of the Dark Triad (Average Transformation)\n\n')
plt.legend(loc='best')
plt.show()
```

## Cronbach Alpha Transformation

```python
cron_alpha = [0.83, 0.90, 0.87, 0.74, 0.77, 0.81, 0.83, 0.89, 0.64, 0.57, 0.84, 0.55]
```

```python
hon = np.zeros((493,12))
dis = np.zeros((493,12))
for i in range(12):
  hon[:,i] = honest.iloc[:,i]*cron_alpha[i]
  dis[:,i] = dishonest.iloc[:,i]*cron_alpha[i]
hon_alpha = pd.DataFrame(hon, columns = e[:12])
dis_alpha = pd.DataFrame(dis, columns = e[:12])
```

```python
x = hon_alpha.loc[:, model_dict['PSY']]
y = hon_alpha.loc[:, model_dict['NAR']]
z = hon_alpha.loc[:, model_dict['MAC']]

coord_alpha = np.array([x.mean(axis=1),y.mean(axis=1),z.mean(axis=1)]).T
```

```python
x = dis_alpha.loc[:, model_dict['PSY']]
y = dis_alpha.loc[:, model_dict['NAR']]
z = dis_alpha.loc[:, model_dict['MAC']]

coord_alpha1 = np.array([x.mean(axis=1),y.mean(axis=1),z.mean(axis=1)]).T
```

```python
from mpl_toolkits import mplot3d

x = coord_alpha[:,0]
y = coord_alpha[:,1]
z = coord_alpha[:,2]

x1 = coord_alpha1[:,0]
y1 = coord_alpha1[:,1]
z1 = coord_alpha1[:,2]

fig = plt.figure(figsize = (10, 10))
ax = plt.axes(projection ="3d")

ax.scatter3D(x, y, z, c='b', marker='o', label='Honest')
ax.scatter3D(x1, y1, z1, c='r', marker='o', label='Disonest')

ax.set_xlabel('Psychopathy')
ax.set_ylabel('Narcissism')
ax.set_zlabel('Machiavellianism')
plt.title('Components of the Dark Triad (Cronbach Alpha Transformation)\n\n')
plt.legend(loc='best')
plt.show()
```

## TF-IDF Transformation

```python
honest = np.array(honest)
dishonest = np.array(dishonest)
```

```python
tf = np.zeros((honest.shape[0], honest.shape[1]))

for column in range(honest.shape[1]):
  for row in range(honest.shape[0]):
    tf[row][column] = (np.sum(honest[:,column]==honest[row][column])/(honest.shape[0])) * np.log10(honest.shape[1]/(np.sum(honest==honest[row][column])))
```

```python
tf1 = np.zeros((honest.shape[0], honest.shape[1]))

for column in range(honest.shape[1]):
  for row in range(honest.shape[0]):
    tf1[row][column] = (np.sum(dishonest[:,column]==dishonest[row][column])/(dishonest.shape[0])) * np.log10(dishonest.shape[1]/(np.sum(dishonest==dishonest[row][column])))
```

```python
tfidf_honest = pd.DataFrame(tf,columns=e[0:12])
```

```python
tfidf_dishonest = pd.DataFrame(tf1,columns=e[0:12])
```

```python
x = tfidf_honest.loc[:, model_dict['PSY']]
y = tfidf_honest.loc[:, model_dict['NAR']]
z = tfidf_honest.loc[:, model_dict['MAC']]

coord = np.array([x.mean(axis=1),y.mean(axis=1),z.mean(axis=1)]).T
```

```python
x = tfidf_dishonest.loc[:, model_dict['PSY']]
y = tfidf_dishonest.loc[:, model_dict['NAR']]
z = tfidf_dishonest.loc[:, model_dict['MAC']]

coord1 = np.array([x.mean(axis=1),y.mean(axis=1),z.mean(axis=1)]).T
```

```python
from mpl_toolkits import mplot3d

x = coord[:,0]
y = coord[:,1]
z = coord[:,2]

x1 = coord1[:,0]
y1 = coord1[:,1]
z1 = coord1[:,2]

fig = plt.figure(figsize = (10, 10))
ax = plt.axes(projection ="3d")

ax.scatter3D(x, y, z, c='b', marker='o', label='Honest')
ax.scatter3D(x1, y1, z1, c='r', marker='o', label='Disonest')

ax.set_xlabel('Psychopathy')
ax.set_ylabel('Narcissism')
ax.set_zlabel('Machiavellianism')
plt.title('Components of the Dark Triad (TF-IDF Transformation)\n\n')
plt.legend(loc='best')
plt.show()
```

```python
tfidf = pd.concat([tfidf_honest,tfidf_dishonest]).reset_index(drop=True)
tfidf['CONDITION'] = Y
```

```python
honest = tfidf[tfidf['CONDITION'] == 'H']
false = tfidf[tfidf['CONDITION'] == 'D']

X_train_h, X_test_h, y_train_h, y_test_h = train_test_split(
        honest.drop(columns = 'CONDITION'),
        honest['CONDITION'], train_size=0.9, shuffle=False)

X_train_f, X_test_f, y_train_f, y_test_f = train_test_split(
        false.drop(columns = 'CONDITION'),
        false['CONDITION'], train_size=0.9, shuffle=False)

df_train = pd.DataFrame(X_train_h.append(X_train_f))
df_train['CONDITION'] = y_train_h.append(y_train_f)

df_test = pd.DataFrame(X_test_h.append(X_test_f))
df_test['CONDITION'] = y_test_h.append(y_test_f)

df_train.to_csv('tfidf_train_dataset.csv', index=False)
df_test.to_csv('tfidf_test_dataset.csv', index=False)

print('Percentage of honest responses in train:')
print(str(round(len(df_train[df_train['CONDITION'] == 'H']) / len(df_train), 2)))

print('Percentage of honest responses in test:')
print(str(round(len(df_test[df_test['CONDITION'] == 'H']) / len(df_test), 2)))
```
